## NFL Offense Data

```python
import pandas as pd
from glob import glob
```

```python
week3 = pd.read_csv("nfloffenseweek3.csv")
week4 = pd.read_csv("nfloffenseweek4.csv")
week5 = pd.read_csv("nfloffenseweek5.csv")
```

1. Import the NFL week 3, 4, and 5 data and add a week column to dataset and set the value to 3, 4, or 5 for all rows from that week.

- The 3 below is each week seperate

```python
week3_column = week3.assign(week=3)
week3_column
```

```python
week4_column = week4.assign(week=4)
week4_column
```

```python
week5_column = week5.assign(week=5)
week5_column
```

- combined all data to create a big dataframe with the different weeks added 

```python
all_data_weeks = pd.concat([week3_column, week4_column, week5_column], ignore_index=True)
all_data_weeks
```

2. Combine the weekly dataframes by adding the dataframes together row-wise.

```python
stacked_files = sorted(glob('nfloffenseweek*.csv'))
stacked_files
```

```python
pd.concat((pd.read_csv(file) for file in stacked_files), ignore_index=True)
```

3. Make up 1 interesting question about this dataset. Answer it 

- Which team had the most amount of total yards in one week?

```python
total = all_data_weeks.loc[ : ,["team", "total_yards","week"]].sort_values("total_yards", ascending=False)
total.head(1)
```
